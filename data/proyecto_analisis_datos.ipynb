```python
#identificar cuales generos y artistas generan mas ingresos y quienes son los mejores clientes

relaciones= []
for t in df_tablas["name"]:
  fk= pd.read_sql_query(f"PRAGMA Foreign_Key_list({t})",conn)
  if not fk.empty:
    fk["tabla_origen"]=t
    relaciones.append(fk)

```

```python
print(relaciones)
```

```python
df_rel= pd.concat(relaciones,ignore_index=True) if relaciones else pd.DataFrame()
```

```python
if not df_rel.empty:
  df_rel = df_rel.rename(columns={
      "table":"tabla_destino",
      "from":"columna_origen",
      "to":"columna_destino"
      })[["tabla_origen","tabla_destino","columna_origen","columna_destino"]]
```

```python
q1_generos="""SELECT g.Name AS Genero, SUM(il.UnitPrice * il.Quantity) AS Ingresos
FROM InvoiceLine il
JOIN Track t ON il.TrackId = t.TrackId
JOIN Genre g ON t.GenreId = g.GenreId
GROUP BY Genero
ORDER BY Ingresos DESC
"""
df_generos_ingresos= pd.read_sql_query(q1_generos,conn)
df_generos_ingresos
```

```python
#identificar cual artistas genera mas ingresos y quienes son los mejores clientes
q1_artista="""SELECT a.Name AS Artista, SUM(il.UnitPrice * il.Quantity) AS Ingresos
FROM InvoiceLine il
JOIN Track t ON il.TrackId = t.TrackId
JOIN Album al ON t.AlbumId = al.AlbumId
JOIN Artist a ON al.ArtistId = a.ArtistId
GROUP BY Artista
ORDER BY Ingresos DESC
LIMIT 10;
"""
df_artista_ingresos= pd.read_sql_query(q1_artista,conn)
df_artista_ingresos
```

```python
#mejores clientes

q3_clientes="""SELECT c.FirstName || ' ' || c.LastName AS Cliente, SUM(i.Total) AS Total_Compras
FROM Customer c
JOIN Invoice i ON c.CustomerId = i.CustomerId
GROUP BY Cliente
ORDER BY Total_Compras DESC
LIMIT 10;
"""
df_Cliente_ingresos= pd.read_sql_query(q3_clientes,conn)
df_Cliente_ingresos
```

```python
#union de las 3 consultas en una sola

q_master = """SELECT
              i.InvoiceId AS Factura,
              i.InvoiceDate AS Fecha_Factura,
              c.FirstName || ' ' || c.LastName AS Cliente,
              c.Country AS Pais,
              ar.Name AS Artista,
              g.Name AS Genero,
              t.Name AS Cancion,
              il.UnitPrice AS Precio_Unitario,
              il.Quantity AS Cantidad,
              (il.UnitPrice * il.Quantity) AS Total_Linea
          FROM InvoiceLine il
          JOIN Invoice i ON i.InvoiceId=il.InvoiceId
          JOIN Customer c ON c.CustomerId=i.CustomerId
          JOIN Track t ON t.TrackId=il.TrackId
          JOIN Genre g ON g.GenreId=t.GenreId
          JOIN Album al ON al.AlbumId=t.AlbumId
          JOIN Artist ar ON ar.ArtistId=al.ArtistId;
          """
df_master= pd.read_sql_query(q_master,conn)
df_master.head()
```

```python
#tipos, nulos, y duplicados
df_master.info()
print("\nNulos por columnas:\n", df_master.isnull().sum())

#Conversion de tipo de dato a fecha
df_master['Fecha_Factura'] = pd.to_datetime(df_master['Fecha_Factura'])

#extraer dia,mes,año para hacer las consultas facilmente
df_master['Dia'] = df_master['Fecha_Factura'].dt.day
df_master['Mes'] = df_master['Fecha_Factura'].dt.month
df_master['Año'] = df_master['Fecha_Factura'].dt.year

#Numero de duplicados exactos si los hubiera
dupli= df_master.duplicated().sum()
print(f"\nDuplicados exactos : {dupli}")

# Eliminamos duplicados si existen
df= df_master.drop_duplicates()
print("Duplicados eliminados. total de filas ahora: ", df.shape[0])

#ahora si alguna fecha no tiene el formato esperado
#vamos a decirle que nos arroje un error NaT
df_master['Fecha_Factura'] = pd.to_datetime(df_master['Fecha_Factura'], errors='coerce')
```

```python
# ahora vamos a convertir los tipo string,
#borramos los espacios, y pondremos la primera
#letra en mayuscula

columnas_string=["Cliente","Pais","Artista","Genero","Cancion"]
for columna in columnas_string:
  df_master[columna]= df_master[columna].astype("string").str.strip().str.title()
```

```python
# ahora vamos a convertir los tipo int,

columnas_int=["Factura","Cantidad","Precio_Unitario","Total_Linea"]
for columna in columnas_int:
  df_master[columna]= pd.to_numeric(df_master[columna], errors='coerce' )
```

```python
columnas_categoria = ["Genero","Artista","Pais"]
for columna in columnas_categoria:
  df_master[columna] = df_master[columna].astype("category")
```

```python
#graficamos
import matplotlib.pyplot as plt
# 1. Agrupamos y calculamos el total general
data_genero = df_master.groupby("Genero")["Total_Linea"].sum().sort_values(ascending=False)
total_general = data_genero.sum()

# 2. Creamos el gráfico
ax = data_genero.plot(kind="bar", figsize=(12, 6), color=['skyblue', 'salmon'])

# 3. Iteramos para poner los porcentajes
for p in ax.patches:
    # Calculamos el % (valor de la barra / total)
    porcentaje = f'{(100 * p.get_height() / total_general):.1f}%'

    # Anotamos el texto sobre la barra
    ax.annotate(porcentaje,
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom',
                xytext=(0, 5),
                textcoords='offset points',
                fontweight='bold')

# Toques finales
plt.title("Distribución porcentual por Género")

```
